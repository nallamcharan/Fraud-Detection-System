```python
from flask import Flask, render_template, request
import joblib
import pandas as pd

app = Flask(__name__)

# Load trained pipeline model
model1 = joblib.load("fraud_model.pkl")

@app.route("/")
def home():
    return render_template("home.html")

@app.route("/", methods=["POST","GET"])
def predict():

    # Get form data
    amount = float(request.form["amount"])
    age = float(request.form["age"])
    merchant = request.form["merchant"]
    zipcodeori = float(request.form["zipcodeOri"]) #city name /zipcode
    transaction_time = float(request.form["time"])

    # Create DataFrame (IMPORTANT: column names must match training)
    input_data = pd.DataFrame([{
        "amount": amount,
        "age": age,
        "merchant": merchant,
        "zipcodeOri": zipcodeori,
        "transaction_time": transaction_time
    }])

    # Prediction
    prediction = model1.predict(input_data)[0]
    probability = model1.predict_proba(input_data)[0][1]

    if prediction == 1:
        result_text = "⚠ Fraudulent Transaction Detected!"
    else:
        result_text = "✅ Legitimate Transaction"

    return render_template(
        "after.html",
        amount=amount,
        age=age,
        merchant=merchant,
        zipcodeOri=zipcodeori,
        transaction_time=transaction_time,
        prediction_result=result_text,
        probability=round(probability * 100, 2)
    )

if __name__ == "__main__":
    app.run(debug=True,use_reloader=False)
```
